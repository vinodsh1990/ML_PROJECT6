```python
import requests
from bs4 import BeautifulSoup
```

```python
# requesting the websited 
url = 'https://www.bbc.com'
```

```python
response = requests.get(url)
response
```

```python
# parshing the HTML
soup = BeautifulSoup(response.content, "html.parser")
soup
```

```python
# extract the news headlines from the HTML
headlines = soup.find_all('h2')
print(headlines)  # see the headlines
```

```python
for headline in headlines:
    print(headline.text)
```

```python
# see the above code but in new lines as
for headline in headlines:
    print(headline.text, '\n')
```

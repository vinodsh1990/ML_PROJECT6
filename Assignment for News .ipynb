```python
import requests
from bs4 import BeautifulSoup
```

```python
# requesting the websited 
urls = ['https://www.bbc.com/sport', 
       'https://www.bbc.com/travel', 
       'https://www.bbc.com/arts', 
       'https://www.bbc.com/culture', 
       'https://www.bbc.com/business']
```

```python
for url in urls:
    response = requests.get(url)    
```

```python
response
```

```python
# parshing the HTML
soup = BeautifulSoup(response.content, "html.parser")
soup
```

```python
# extract the news headlines from the HTML
headlines = soup.find_all('h2')
print(headlines)  # see the headlines
```

```python
for headline in headlines:
    print(headline.text)
```

```python
# see the above code but in new lines as
for headline in headlines:
    print(headline.text, '\n')
```
